```python
class Food:
    def __init__(self,name,category,origin):
        self.name=name
        self.category=category
        self.origin=origin
    def __str__(self):
        return (f'{self.name} is a {self.category} food from {self.origin}.')

class NutritionalValue:
    def __init__(self,calories):
        self.calories=calories
    def is_healthy(self):
        return self.calories<500

class Fruit(Food):
    def __init__(self,name,origin):
        super().__init__(name,'Fruit',origin)

class Vegetable(Food,NutritionalValue):
    def __init__(self,name,origin,calories):
        super().__init__(name,'Vegetable',origin)
        NutritionalValue.__init__(self,calories)

class Meat(Food,NutritionalValue):
    def __init__(self,name,origin,calories):
        super().__init__(name,'Meat',origin)
        NutritionalValue.__init__(self,calories)

steak =Meat('Steak','Argentina',600)
print(steak)
print(isinstance(steak,Food))
print(isinstance(steak,NutritionalValue))
print(steak.calories==600)
print(not steak.is_healthy())
carrot=Vegetable('Carrot','USA',50)
print(carrot)
print(isinstance(carrot,Food))
print(isinstance(carrot,NutritionalValue))
apple=Fruit('Apple','France')
print(apple)
print(isinstance(apple,Food))
rice=Food('Rice','Grain','China')
print(rice)
```

```python
class Shape:
    def __init__(self,volume):
        self.__volume=volume

    def get_volume(self):
        return self.__volume

    def set_volume(self,volume):
        self.__volume=volume

    def __str__(self):
        return f'Shape has volume {self.get_volume()} units.'

    def __add__(self,other):
        return self.get_volume()+other.get_volume()

    def __sub__(self,other):
        return self.get_volume()-other.get_volume()

    def __iadd__(self,other):
        self.set_volume(self.get_volume()+other.get_volume())
        return self

    def __isub__(self,other):
        self.set_volume(self.get_volume()-other.get_volume())
        return self

    def __lt__(self,other):
        return self.get_volume()<other.get_volume()

    def __repr__(self):
        return str(self.get_volume())

my_shape=Shape(10)
# print(my_shape)
zone_1=Shape(10)
zone_2=Shape(5)
# print(zone_1)
print(zone_1+zone_2)
print(zone_1-zone_2)
zone_1+=zone_2
print(zone_1.get_volume())

import random
random.seed(42)
shapes=list()
for _ in range(5):
    shapes.append(Shape(random.randint(1,10)))
shapes.sort()
print(shapes)
shapes.__repr__()
```

```python
class Rock:
    def __init__(self,name):
        self._name=name

    def hardness(self):
        return 'undefined'

    def __str__(self):
        return f'The rock {self._name} is {self.hardness()}'

class Igneous(Rock):
    def hardness(self):
        return 'harder'

class Metamorphic(Rock):
    def hardness(self):
        return 'hardest'

class Sedimentary(Rock):
    def hardness(self):
        return 'softest'

rocks=[Igneous('Granite'),Sedimentary('Chalk'),Metamorphic('Marble')]
for rock in rocks:
    print(rock,type(rock))
    if type(rock) == Igneous:
        print('harder' in rock.__str__())
    elif type(rock) == Metamorphic:      
        print('hardest' in rock.__str__())
    elif type(rock) == Sedimentary:      
        print('softest' in rock.__str__())
```

```python
from random import choice
colours=['white','grey','brown','black','tan']
class Saddle:
    def __init__(self,colour,style='English'):
        self._colour=colour
        self._style=style

class Leg:
    def __init__(self,main_colour,fetlock_colour='black'):
        self._main_colour=main_colour
        self._fetlock_colour=fetlock_colour

class Horse:
    def __init__(self,name,saddle):
        self._name=name
        self._saddle=saddle
        self._legs=list()
        for i in range(4):
            self._legs.append(Leg(choice(colours)))

    def __str__(self):
        return f'{self._name} has {len(self._legs)} legs and is wearing an {self._saddle._style} saddle.'

deedee=Horse('DeeDee',Saddle('tan'))
```

```python
class Mohs:
    def __init__(self,hardness):
        self._hardness=hardness

    @property
    def hardness(self):
        return self._hardness

    def __str__(self):
        return f'object of class {self.__class__.__name__} has hardness {self._hardness}'

    def __lt__(self,other):
        return self.hardness<other.hardness

    def __gt__(self,other):
        return self.hardness>other.hardness

    def __eq__(self,other):
        return self.hardness==other.hardness

class Mineral(Mohs):
    def __init__(self,name,hardness):
        super().__init__(hardness)
        self._name=name

    @property
    def name(self):
        return self._name

class Drill(Mohs):
    __STANDARD_MATERIAL='Hardened steel'
    __DEFAULT_HARDNESS=7.5
    def __init__(self,material,hardness):
        super().__init__(hardness)
        self._material=material

    @property
    def material(self):
        return self._material

    @classmethod
    def default_bit(cls):
        return cls(cls.__STANDARD_MATERIAL,cls.__DEFAULT_HARDNESS)

    def can_drill(self,mineral):
        ''' 
        Compares the hardness of the drill bit to the hardness of the material
        and returns True if the drill is harder than the material. Otherwise
        False.
        '''
        return self.hardness>mineral.hardness

mohs_scale = [
    Mineral ('talc', 1),
    Mineral ('gypsum', 2),
    Mineral ('calcite', 3),
    Mineral ('fluorite', 4),
    Mineral ('apatite', 5),
    Mineral ('feldspar', 6),
    Mineral ('quartz', 7),
    Mineral ('topaz', 8),
    Mineral ('corundum', 9),
    Mineral ('diamond', 10),
]

drill=Drill.default_bit()
for m in mohs_scale:
    if drill.can_drill(m):
        print(f'{drill.material} is harder than {m.name}.')
    else:
        print(f'{drill.material} is softer than {m.name}.')
```
