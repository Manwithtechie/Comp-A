### Doubly Linked Lists

```python
class Node:
    def __init__(self,data):
        self.data=data
        self.prev=None
        self.next=None

class DoublyLinkedList:
    def __init__(self):
        self.head=None
        self.tail=None

    def traverse(self):
        cur_node=self.head
        if self.head==None:
            print('List is empty')
            return
        print('None',end='')
        while cur_node!=None:
            print(f'<--{cur_node.data}-->',end='')
            cur_node=cur_node.next
        print('None')

    def append(self,data):
        new_node=Node(data)
        if self.head==None:
            self.head=self.tail=new_node
            self.head.prev=None
            self.tail.next=None
        else:
            self.tail.next=new_node
            new_node.prev=self.tail
            self.tail=new_node
            self.tail.next=None

    def prepend(self,data):
        new_node=Node(data)
        if self.head==None:
            self.head=self.tail=new_node
            self.head.prev=None
            self.tail.next=None
        else:
            self.head.prev=new_node
            new_node.next=self.head
            self.head=new_node
            self.head.prev=None

    def delete_first(self):
        if self.head==None:
            return
        else:
            if self.head!=self.tail:
                self.head=self.head.next
                self.head.prev=None
            else:
                self.head=None
                self.tail=None
            

dllist=DoublyLinkedList()
dllist.append(1)
dllist.append(2)
dllist.append(3)
dllist.append(4)
dllist.traverse()
dllist.prepend(12)
dllist.prepend(13)
dllist.prepend(14)
dllist.prepend(15)
dllist.traverse()
dllist.delete_first()
dllist.traverse()
```

```python
class Node:
    def __init__(self,data):
        self._data=data
        self._next=None

    @property
    def data(self):
        return self._data

    @data.setter
    def data(self,value):
        self._data=value

    @property
    def next(self):
        return self._next

    @next.setter
    def next(self,node):
        self._next=node

    def __str__(self):
        return f'Node({self._data})'

    def __repr__(self):
        return f'Node({repr(self._data)})'

    # Factory Pattern
    @classmethod
    def from_list(cls,data_list):
        """Factory method to build a chain of nodes from a Python list."""
        if not data_list:
            return None
        head=cls(data_list[0])
        cur_node=head
        for item in data_list[1:]:
            cur_node._next=cls(item)
            cur_node=cur_node._next
        return head

class LinkedList:
    def __init__(self):
        self.head=None

    def traverse(self):
        cur_node=self.head
        while cur_node!=None:
            print(cur_node.data,end='-->')
            cur_node=cur_node.next
        print('None')

    def __add__(self,other):
        """Merge anohter LinkedList to this one."""
        if self.head is None:
            self.head=other.head
        cur_node=self.head
        while cur_node.next:
            cur_node=cur_node.next
        cur_node.next=other.head
        return self

    def remove_duplicates(self):
        """Remove duplicate values in the list."""
        cur_node=self.head
        seen=set()
        prev=None
        while cur_node:
            if cur_node.data in seen:
                prev.next=cur_node.next
            else:
                seen.add(cur_node.data)
                prev=cur_node
            cur_node=cur_node.next

    def count_occurences(self,value):
        """Count how many times 'value' appears."""
        cur_node=self.head
        cnt=0
        while cur_node:
            if cur_node.data==value:
                cnt+=1
            cur_node=cur_node.next
        return cnt

    def is_palindrome(self):
        """Check whether the list is a palindrome."""
        cur_node=self.head
        ll=[]
        while cur_node:
            ll.append(cur_node.data)
            cur_node=cur_node.next
        return ll==ll[::-1]

    def reverse_ll(self):
        """Reverse the Linked List"""
        prev=None
        cur=self.head
        while cur:
            nxt=cur.next
            cur.next=prev
            prev=cur
            cur=nxt
        self.head=prev

#empty list
ll1=LinkedList()
ll1.head=Node.from_list([5,6,3])
ll1.traverse()
ll2=LinkedList()
ll2.head=Node.from_list([8,4,2])
ll2.traverse()
ll=ll1+ll2
ll.traverse()
ll3=LinkedList()
ll3.head=Node.from_list([1,2,3,4,3,6,2,4])
ll3.traverse()
ll3.remove_duplicates()
ll3.traverse()
ll4=LinkedList()
ll4.head=Node.from_list([1,2,1,4,1,6,2,4])
ll4.traverse()
print('Number of occurences of 1:',ll4.count_occurences(1))
ll5=LinkedList()
ll5.head=Node.from_list([1,2,1,2,1,2,1])
print(ll5.is_palindrome())
ll6=LinkedList()
ll6.head=Node.from_list(['John','Danise','Kelly','David'])
ll6.reverse_ll()
ll6.traverse()
```
