## 4. Using SLL

```python
class Node:
    def __init__(self,data):
        self.data=data
        self.next=None

class Stack:
    def __init__(self):
        self.head=None

    def isempty(self):
        if self.head==None:
            return True
        else:
            return False

    def push(self,data):
        if self.head==None:
            self.head=Node(data)

        else:
            newnode=Node(data)
            newnode.next=self.head
            self.head=newnode

    def pop(self):
        if self.isempty():
            return None
        else:
            poppednode=self.head
            self.head=self.head.next
            poppednode.next=None
            return poppednode.data

    def peek(self):
        if self.isempty():
            return None
        else:
            return self.head.data

    def display(self):
        iternode=self.head
        if self.isempty():
            print("Stack Underflow")
        else:
            while iternode!=None:
                print(iternode.data,'->',end=' ')
                iternode=iternode.next
            return

stack=Stack()
stack.push(11)
stack.push(22)
stack.push(33)
stack.push(44)

stack.display()

print("\nTop element is: ",stack.peek())

stack.pop()
stack.pop()

stack.display()

print("\nTop element is: ",stack.peek())
```

```python
class Queue:
    def __init__(self):
        self.elements=[]

    def enqueue(self,data):
        self.elements.append(data)
        return data

    def dequeue(self):
        return self.elements.pop(0)

    def rear(self):
        return self.elements[-1]

    def front(self):
        return self.elements[0]

    def is_empty(self):
        return len(self.elements)==0

    def display(self):
        for i in self.elements:
            print(i,end=' ')    
        print()

queue=Queue()
print(queue.is_empty())

queue.enqueue(2)
queue.enqueue(4)
queue.enqueue(6)
queue.enqueue(8)
queue.enqueue(10)

queue.display()

print(queue.is_empty())

print('Front: ',queue.front())
print('Rear: ',queue.rear())

queue.dequeue()

queue.display()
print('Front: ',queue.front())
print('Rear: ',queue.rear())

queue.dequeue()
queue.dequeue()
queue.dequeue()
queue.dequeue()

print(queue.is_empty())
```

## 4. Using SLL

```python
class Node:
    def __init__(self,data):
        self.data=data
        self.next=None

class Queue:
    def __init__(self):
        self.front=None
        self.rear=None

    def is_empty(self):
        return self.front==None

    def enqueue_operation(self,item):
        current_node=Node(item)
        if self.rear==None:
            self.front=self.rear=current_node
            return
        self.rear.next=current_node
        self.rear=current_node

    def dequeue_operation(self):
        if self.is_empty():
            return
        current_node=self.front
        self.front=current_node.next

        if(self.front==None):
            self.rear=None

q=Queue()
q.enqueue_operation(2)
q.enqueue_operation(4)
q.dequeue_operation()
q.dequeue_operation()
q.enqueue_operation(6)
q.enqueue_operation(8)
q.enqueue_operation(10)
q.dequeue_operation()
print('Queue Front: ',q.front.data)
print('Queue Rear: ',q.rear.data)
```

```python
class Node:
    def __init__(self,data):
        self.data=data
        self.next=None

class Stack:
    def __init__(self):
        self.head=None

    def push(self,data):
        new_node=Node(data)
        if self.head==None:
            self.head=new_node
        else:
            new_node.next=self.head
            self.head=new_node

    def pop(self):
        if self.head is None:
            return None
        popped=self.head.data
        self.head=self.head.next
        return popped

    def top(self):
        if self.head is None:
            return None
        return self.head.data

    def is_empty(self):
        return self.head is None

    def display(self):
        cur=self.head
        while cur:
            print(cur.data,end='')
            cur=cur.next

def contains_duplicate(expr):
    s=Stack()
    for i in expr:
        if i != ')':
            s.push(i)
        else:
            if s.top()=='(':
                return True
                
            while not s.is_empty() and s.top()!='(':
                s.pop()
            s.pop()
    return False

st=input('Enter a sequence: ')
if contains_duplicate(st):
    print('Found Duplicate')
else:
    print('No Duplicate Found')
```

```python
class Node:
    def __init__(self,data):
        self.data=data
        self.next=None
class Queue:
    def __init__(self):
        self.front=None
        self.rear=None

    def enqueue(self,value):
        new_node=Node(value)
        if self.front==None:
            self.front=new_node
            self.rear=new_node
        else:
            self.rear.next=new_node
            self.rear=new_node

    def dequeue(self):
        if self.front==None:
            return None
        elif self.front==self.rear:
            front_element=self.front.data
            self.front=None
            self.rear=None
            return front_element
        else:
            front_element=self.front.data
            self.front=self.front.next
            return front_element

    def front_element(self):
        if self.front is None:
            return None
        return self.front.data

    def count_elements(self):
        cur=self.front
        cnt=0
        while cur:
            cur=cur.next
            cnt+=1
        return cnt

    def is_empty(self):
        return self.rear==None

    def display(self):
        cur=self.front
        while cur:
            print(cur.data,end='->')
            cur=cur.next
        print()

class Stack:
    def __init__(self):
        self.q1=Queue()
        self.q2=Queue()

    def push(self,data):
        self.q1.enqueue(data)

    def pop(self):
        if self.q1.is_empty():
            return None
        while self.q1.count_elements()>1:
            self.q2.enqueue(self.q1.dequeue())
        popped=self.q1.dequeue()
        self.q1, self.q2 = self.q2, self.q1
        return popped

    def size(self):
        return self.q1.count_elements()

    def is_empty(self):
        return self.q1.is_empty()

    def display(self):
        self.q1.display()

s=Stack()
s.push(1)
s.push(2)
s.push(3)

s.display()             
print(s.pop())          
s.display()             
print(s.size())         
print(s.is_empty())
print(s.pop())
print(s.pop())
print(s.is_empty())
```

```python
class Node:
    def __init__(self,data):
        self.data=data
        self.next=None

class Stack:
    def __init__(self):
        self.head=None

    def push(self,data):
        new_node=Node(data)
        if self.head==None:
            self.head=new_node
        else:
            new_node.next=self.head
            self.head=new_node

    def pop(self):
        if self.head is None:
            return None
        popped=self.head.data
        self.head=self.head.next
        return popped

    def top(self):
        if self.head is None:
            return None
        return self.head.data

    def count_elements(self):
        cur=self.head
        cnt=0
        while cur:
            cur=cur.next
            cnt+=1
        return cnt
    def is_empty(self):
        return self.head is None

    def display(self):
        cur=self.head
        while cur:
            print(cur.data,end='')
            cur=cur.next

class Queue:
    def __init__(self):
        self.inbox=Stack()
        self.outbox=Stack()

    def enqueue(self,value):
        self.inbox.push(value)

    def dequeue(self):
        if self.outbox.is_empty():
            while not self.inbox.is_empty():
                self.outbox.push(self.inbox.pop())
        return self.outbox.pop()

    def count_elements(self):
        return self.inbox.count_elements()+self.outbox.count_elements()

    def is_empty(self):
        return self.inbox.is_empty() and self.outbox.is_empty()

q = Queue()
q.enqueue(1)
q.enqueue(2)
q.enqueue(3)

print(q.dequeue())  # 1
print(q.dequeue())  # 2
q.enqueue(4)
print(q.dequeue())  # 3
print(q.dequeue())  # 4
print(q.dequeue())  # None
```

```python
print('--------- USING STACK ---------')

class Node:
    def __init__(self,data):
        self.data=data
        self.next=None

class Stack:
    def __init__(self):
        self.head=None

    def push(self,data):
        new_node=Node(data)
        if self.head==None:
            self.head=new_node
        else:
            new_node.next=self.head
            self.head=new_node

    def pop(self):
        if self.head is None:
            return None
        popped=self.head.data
        self.head=self.head.next
        return popped

    def top(self):
        if self.head is None:
            return None
        return self.head.data

    def is_empty(self):
        return self.head is None

    def display(self):
        cur=self.head
        s=''
        while cur:
            s+=cur.data
            cur=cur.next
        return s

s=Stack()
st=input('Enter a string: ')
for i in st:
    s.push(i)
reversed_st=s.display()
if st==reversed_st:
    print('Yes')
else:
    print('No')


print('--------- USING QUEUE ---------')
class Node:
    def __init__(self,data):
        self.data=data
        self.next=None
class Queue:
    def __init__(self):
        self.front=None
        self.rear=None

    def enqueue(self,value):
        new_node=Node(value)
        if self.front==None:
            self.front=new_node
            self.rear=new_node
        else:
            self.rear.next=new_node
            self.rear=new_node

    def dequeue(self):
        if self.front==None:
            return None
        elif self.front==self.rear:
            front_element=self.front.data
            self.front=None
            self.rear=None
            return front_element
        else:
            front_element=self.front.data
            self.front=self.front.next
            return front_element

    def count_elements(self):
        cur=self.front
        cnt=0
        while cur:
            cur=cur.next
            cnt+=1
        return cnt

    def is_empty(self):
        return self.rear==None

    def display(self):
        cur=self.front
        while cur:
            print(cur.data,end='->')
            cur=cur.next
        print()

q = Queue()
st = input("Enter a string: ")

for ch in st:
    q.enqueue(ch)

queue_string = ''
while not q.is_empty():
    queue_string += q.dequeue()

if st == queue_string:
    print("Yes, palindrome")
else:
    print("No, not palindrome")
```
